# **Filtering out missing data**

```python
data=pd.Series([2,3,9,np.nan,2.1,np.nan])
data
```

```python
data.dropna()
```

```python
data[data.notna()]
```

```python
data=pd.DataFrame([[2,2.7,8,7.6,9],[np.nan,np.nan,np.nan],[3,np.nan,np.nan]])
data
```

```python
data.dropna()
```

```python
data.dropna(how="all")
```

```python
data[4]=np.nan
data
```

```python
data.dropna(axis="columns",how="all")
```

```python
df=pd.DataFrame(np.random.standard_normal((5,7)))
df
```

```python
df.iloc[:3,1]=np.nan
df.iloc[:2,3]=np.nan
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

# **Data Transformation**

```python
#removing duplicates
data=pd.DataFrame({"X":["One","Two"] *3 + ["Two"] , "Y":[3,3,7,7,4,4,1]})
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data["Z"]=range(7)
data
```

```python
data.drop_duplicates(subset=["Y"])
```

```python
data.drop_duplicates(["X","Y"],keep="last")
```

```python
data=pd.DataFrame({"Fruits":["apple","kiwi","berry","Guava","Strawberry","cherry","mango"],
                  "Ounces":   [2,4,5,6,7,8,2]})
data
```

```python
Fruits_to_Killos={
    "apple":1.4,
    "kiwi" :2.6,
    "berry":2.2,
    "Guava":1.2,
    "Strawberry":1.9,
    "Cheery":4.0,
    "mango":2.0,
}
```

```python
data["killos"]=data["Fruits"].map(Fruits_to_Killos)
data
```

```python
#renaming axis indexes
data=pd.DataFrame(np.arange(12).reshape((4,3)),
                  index=["Colorado","ohio","Utah","NYC"],
                  columns=["one","two","three"])
data
```

```python
def transform(x):
  return x[:4].upper()
```

```python
data.index.map(transform)
```

```python
data.index=data.index.map(transform)
data
```
